#  Definition of variables #
UNITID:- Unit id is an unique identification number for each university 

INSTNM:-It's the name of the institution,string data ype

CITY:-It's a name of the city,where institution is located,string data type

STABBR:-It's a code of state where institution is located,string data type 

ZIP:-It's the area's code ,where institutions are located,string data type 

ADM_RATE:-It's an overall admission rate of the institution,float data type 

UGDS:-It's an enrollment of undergraduate students seeking degree or certificates 

TutionFee:-It's institutions tution fee,integer in value

MN_EARN_WNE_P10:-It is an mean earnings of students who are working and not enrolled after 10 years of entry
